# Deterministic onshore wind NPV

Calculate a single onshore wind NPV case using the shared deterministic electricity source code. The notebook displays the representative input values and the processed NPV outputs without duplicating the model formulas.


```python
from pathlib import Path
import sys

import pandas as pd

PROJECT_ROOT = Path.cwd()
while PROJECT_ROOT != PROJECT_ROOT.parent and not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.insert(0, str(SRC_DIR))

from electricity.electricity_npv_deterministic import calculate_deterministic_electricity_result
```

```python
TECHNOLOGY = "wind_onshore"

result = calculate_deterministic_electricity_result(TECHNOLOGY)
value = {key: values[0] for key, values in result.items()}

UNITS = {
    "annual_output_mwh": "MWh/year",
    "full_load_hours_per_year": "h/year",
    "capex_eur_per_kw": "EUR/kW",
    "fixed_opex_eur_per_kw_year": "EUR/kW/year",
    "variable_opex_eur_per_mwh": "EUR/MWh_e",
    "fuel_consumption_mwh_th_per_mwh_e": "MWh_th/MWh_e",
    "emissions_tco2_per_mwh_e": "tCO2/MWh_e",
    "fuel_price_eur_per_mwh_th": "EUR/MWh_th",
    "electricity_price_eur_per_mwh": "EUR/MWh_e",
    "carbon_price_eur_per_t": "EUR/tCO2",
    "capacity_mw": "MW",
    "capacity_kw": "kW",
    "initial_capex_eur": "EUR",
    "annual_revenue_eur": "EUR/year",
    "annual_fixed_opex_eur": "EUR/year",
    "annual_variable_opex_eur": "EUR/year",
    "annual_fuel_cost_eur": "EUR/year",
    "annual_emissions_cost_eur": "EUR/year",
    "annual_net_cash_flow_eur": "EUR/year",
    "npv_eur": "EUR",
    "discounted_lifetime_output_mwh": "MWh",
    "levelized_net_margin_eur_per_mwh": "EUR/MWh",
}

RAW_INPUT_COLUMNS = (
    "annual_output_mwh",
    "full_load_hours_per_year",
    "capex_eur_per_kw",
    "fixed_opex_eur_per_kw_year",
    "variable_opex_eur_per_mwh",
    "fuel_consumption_mwh_th_per_mwh_e",
    "emissions_tco2_per_mwh_e",
    "fuel_price_eur_per_mwh_th",
    "electricity_price_eur_per_mwh",
    "carbon_price_eur_per_t",
)

PROCESSED_OUTPUT_COLUMNS = (
    "capacity_mw",
    "capacity_kw",
    "initial_capex_eur",
    "annual_revenue_eur",
    "annual_fixed_opex_eur",
    "annual_variable_opex_eur",
    "annual_fuel_cost_eur",
    "annual_emissions_cost_eur",
    "annual_net_cash_flow_eur",
    "npv_eur",
    "discounted_lifetime_output_mwh",
    "levelized_net_margin_eur_per_mwh",
)

def build_table(columns):
    return pd.DataFrame(
        {
            "parameter": columns,
            "value": [value[column] for column in columns],
            "unit": [UNITS[column] for column in columns],
        }
    )

summary = pd.DataFrame(
    {
        "metric": (
            "Technology",
            "NPV",
            "NPV",
            "NPV per lifetime output",
            "Annual net cash flow",
            "Lifetime output",
        ),
        "value": (
            value["technology"],
            value["npv_eur"],
            value["npv_eur"] / 1_000_000,
            value["levelized_net_margin_eur_per_mwh"],
            value["annual_net_cash_flow_eur"],
            value["discounted_lifetime_output_mwh"],
        ),
        "unit": (
            "technology key",
            "EUR",
            "MEUR",
            "EUR/MWh",
            "EUR/year",
            "MWh",
        ),
    }
)
raw_inputs = build_table(RAW_INPUT_COLUMNS)
processed_outputs = build_table(PROCESSED_OUTPUT_COLUMNS)
```
